creating dataframe for all the pages

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

final = pd.DataFrame()

for j in range(1, 103):
    webpage = requests.get('https://www.trustpilot.com/categories/restaurants_bars').text
    soup = BeautifulSoup(webpage, 'lxml')

    company = soup.find_all(
        'div',
        class_='paper_paper__EGeEb paper_outline__bqVmn card_card__yyGgu card_noPadding__OOiac styles_wrapper__Jg8fe'
    )

    name, website, res_type, review_counts, location = [], [], [], [], []

    for i in company:
        name.append(i.find('p', class_='typography_heading-xs__osRhC typography_appearance-default__t8iAq').text.strip())

        website.append(i.find('p', class_='typography_body-m__k2UI7 typography_appearance-subtle__PYOVM styles_websiteUrlDisplayed__lSw1A').text.strip())

        res_type.append(i.find('span', class_='typography_body-s__IqDta typography_appearance-default__t8iAq').text.strip())

        review_text = i.find('p', class_='typography_body-m__k2UI7 typography_appearance-subtle__PYOVM styles_ratingText__A2dmB').text.strip()
        review_counts.append(review_text.split('|')[1].split()[0].replace(',', ''))

        location_element = i.find('span', class_='typography_body-m__k2UI7 typography_appearance-subtle__PYOVM styles_metadataItem__DOu6t styles_location__wea8G')
        location.append(location_element.get_text(separator="", strip=True).replace(',', ", ") if location_element else "")

    df = pd.DataFrame({
        'Name': name,
        'Website': website,
        'Restaurant_Type': res_type,
        'Reviews': review_counts,
        'Location': location,
    })
    final = pd.concat([final, df], ignore_index=True)
```
